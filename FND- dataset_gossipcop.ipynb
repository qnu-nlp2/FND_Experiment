```python
import pandas as pd 
import numpy as np 

df = pd.read_csv('gossipcop.csv')
#df.drop('Unnamed: 0', axis=1, inplace=True)
df.head()
```

```python

from sklearn.feature_extraction.text import CountVectorizer

X = df["title"]+df["news_body"]+str(df["count_like"])+str(df["count_share"])+df["user_ids"]
y = df["label"]


vectorizer = CountVectorizer()

vectorizer.fit(X)
transformed = vectorizer.transform(X)
transformed
```

```python
vectorizer.vocabulary_
```

```python
labels = y == "real"
labels = labels.astype(int)
labels
```

```python
### Run your classifier

from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
import pandas as pd 
import numpy as np 
from sklearn.model_selection import cross_val_predict

df = pd.read_csv('gossipcop.csv')
#df.drop('Unnamed: 0', axis=1, inplace=True)
df.head()

from sklearn.feature_extraction.text import CountVectorizer


vectorizer = CountVectorizer()

vectorizer.fit(X)
transformed = vectorizer.transform(X)
#transformed

labels = y == "real"
#labels=y
labels = labels.astype(int)
#labels

X_train, X_test, y_train, y_test = train_test_split(transformed,labels, test_size=0.2,random_state=25)

clf= SVC(kernel="linear")
#svc = svm.SVC (C = 1, kernel = 'linear')
#clf = svc.fit (transformed, labels)
clf.fit(X_train, y_train)
preds = clf.predict(X_test)
#predicted = cross_val_predict(clf,transformed, labels,cv=5)

print ("Accuracy: ")
print (accuracy_score(y_test, preds))
#print((preds == y_test).sum()/len(y_test))
print ("precision: ")
#print (precision_score(labels, predicted, average="weighted"))
print (precision_score(y_test, preds))
print ("recall: ")
print (recall_score(y_test, preds))

print ("f1 ")
print (f1_score(y_test, preds))
```

```python
from sklearn.naive_bayes import GaussianNB

clf2 = GaussianNB()
clf2.fit(X_train.todense(), y_train)
preds = clf2.predict(X_test.todense())
#(preds == y_test).sum()/len(y_test)
print ("Accuracy: ")
print (accuracy_score(y_test, preds))
print ("precision: ")
#print (precision_score(labels, predicted, average="weighted"))
print (precision_score(y_test, preds))
print ("recall: ")
print (recall_score(y_test, preds))

print ("f1 ")
print (f1_score(y_test, preds))
```
